# Challenge: Password Strength Checker

### Problem

Create a program that checks whether a password is strong based on certain criteria. A strong password might include uppercase letters, lowercase letters, numbers, and symbols.

### Tasks

1. Get a password from user input
2. Check for the following:
- Minimum length (e.g., 8 characters)
- Contains at least one uppercase letter
- Contains at least one lowercase letter
- Contains at least one number
- Contains at least one symbol
3. Return a message that rates the password as “Strong,” “Medium,” or “Weak”

```python
def check_password_strength():
    password = input("Enter a password: ")

    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_symbol = any(not char.isalnum() for char in password)
    long_enough = len(password) >= 8

    
    score = sum([has_upper, has_lower, has_digit, has_symbol, long_enough])

    
    if score == 5:
        rating = "Strong"
    elif score == 4:
        rating = "Medium"
    else:
        rating = "Weak"

    
    missing = []
    if not long_enough: missing.append("length (8+ chars)")
    if not has_upper:  missing.append("uppercase letter")
    if not has_lower:  missing.append("lowercase letter")
    if not has_digit:  missing.append("number")
    if not has_symbol: missing.append("symbol")

    
    if missing:
        print(f"Result: {rating} (missing {', '.join(missing)})")
    else:
        print(f"Result: {rating}")
```

```python
check_password_strength()
```
