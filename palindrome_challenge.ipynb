# Challenge: Is It a Palindrome?

### Problem

Write a function that checks whether a given string is a palindrome â€” a word or phrase that reads the same backward as forward (ignoring capitalization, spaces, and punctuation).

### Tasks

1. Define a function called is_palindrome that accepts a single string.
2. Return True if the string is a palindrome, and False otherwise.
3. Normalize the string by:
3.1 Converting it to lowercase
3.2 Removing spaces (and optionally punctuation)
4. Reverse the normalized string and compare it to the original normalized version.

```python
import re

def is_palindrome(text):
    
    cleaned = re.sub(r'[^a-z0-9]', '', text.lower())
    reversed_text = cleaned[::-1]
    return cleaned == reversed_text
```

```python
print(is_palindrome("racecar"))                         
print(is_palindrome("hello"))                           
print(is_palindrome("A man a plan a canal Panama"))    
```

2. Adapt the function to check if a number is a palindrome

```python
import re

def is_number_palindrome(num):
    text = str(num)  
    cleaned = re.sub(r'[^0-9]', '', text)
    return cleaned == cleaned[::-1]
```

Test if a number is a palindrome

```python
print(is_number_palindrome(321))        
print(is_number_palindrome(12321))        
```

3. Return a message instead of a Boolean (e.g., "Yes, it's a palindrome!")

```python
import re

def is_palindrome_message(text):
    cleaned = re.sub(r'[^a-z0-9]', '', text.lower())
    
    if cleaned == cleaned[::-1]:
        return "Yes, it's a palindrome!"
    else:
        return "No, it's not a palindrome."
```
