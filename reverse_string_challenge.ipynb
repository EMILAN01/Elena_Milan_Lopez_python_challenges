# Challenge: Reverse That String!

### Problem

Write a function that takes a single string as input and returns that string in reverse order.

### Tasks

1. Define a function called reverse_string that accepts one argument (a string).
2. Return the reversed version of the string.



First approach: String slicing

```python
def reverse_string(text):
    
    return text[::-1]
```

Test First approach: String slicing

```python
print(reverse_string("hello"))       
print(reverse_string("Python"))      
print(reverse_string("Elena"))       
```

Second approach: For loop

```python
def reverse_string_loop(text):
    
    reversed_str = ""
    for char in text:
        reversed_str = char + reversed_str  
    return reversed_str
```

Test Second approach: For loop

```python
print(reverse_string_loop("hello"))       
print(reverse_string_loop("Python"))      
print(reverse_string_loop("Elena"))     
```

### Stretch Ideas

Add a feature that ignores whitespace or punctuation
   

```python
import string

def reverse_clean(text):
    
    cleaned = "".join(ch for ch in text if ch not in string.punctuation and ch != " ")
    return cleaned[::-1]

```

Test stretch idea

```python
print(reverse_clean("hello"))       
print(reverse_clean("Python"))      
print(reverse_clean("Elena"))     
```
