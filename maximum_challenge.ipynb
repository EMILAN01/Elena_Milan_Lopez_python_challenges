# Challenge: Find the Max

### Problem

Write a function that takes a list of numbers and returns the maximum value in the list — the largest number.

### Tasks

1. Define a function called find_max that accepts a list of integers or floats.
2. Return the largest number in the list.
3. Do not use the built-in max() function — try to solve it manually using:
- A for loop to check each number
- A variable to track the current maximum

```python
def find_max(numbers):
    
    if not numbers:
        return "Error: The list is empty."
    
    current_max = numbers[0]

    for num in numbers:
        if num > current_max:
            current_max = num

    return current_max
```

Test

```python
print(find_max([4, 9, 1, 17, 2]))       
print(find_max([-5, -9, -2, -12]))      
print(find_max([42]))        
```

### Stretch Ideas

Allow the function to work with mixed positive and negative values

Return both the max value and its index in the list

```python
def find_max_with_index(numbers):
    if not numbers:
        return "Error: The list is empty."
    
    current_max = numbers[0]
    max_index = 0
    
    for i in range(len(numbers)):
        if numbers[i] > current_max:
            current_max = numbers[i]
            max_index = i
    
    return current_max, max_index
```

```python
print(find_max_with_index([4, 9, 1, 17, 2]))       
print(find_max_with_index([-5, -9, -2, -12]))      
print(find_max_with_index([42]))  
```
